```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Definimos la primera funci贸n
def funcion_a(x):
    return 3 * np.sin(np.pi * x)

# Generamos datos de entrenamiento en el intervalo [-1,1]
x_train = np.linspace(-1, 1, 1000)
y_train_a = funcion_a(x_train)

# Definimos la arquitectura de la red neuronal con Sequential
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compilamod el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenamos el modelo para la primera funci贸n
model.fit(x_train, y_train_a, epochs=100, verbose=0)

# Usamos predict para la red
y_pred_a = model.predict(x_train)


# Graficamos los resultados
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x_train, y_train_a, label='Funci贸n a')
plt.plot(x_train, y_pred_a, label='Red Neuronal')
plt.title('Funci贸n a')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Definimos la segunda
def funcion_b(x):
    return 1 + 2 * x + 4 * x**3

# Generamos datos de entrenamiento en el intervalo [-1,1]
x_train = np.linspace(-1, 1, 1000)
y_train_b = funcion_b(x_train)

# Definimos la arquitectura de la red neuronal con Sequential
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compilamos el modelo
model.compile(optimizer='adam', loss='mean_squared_error')


# Lo mismo que arriba para la funci贸n b
model.fit(x_train, y_train_b, epochs=100, verbose=0)

y_pred_b = model.predict(x_train)

# Graficamos los resultados
plt.figure(figsize=(10, 5))


plt.subplot(1, 2, 2)
plt.plot(x_train, y_train_b, label='Funci贸n b')
plt.plot(x_train, y_pred_b, label='Red Neuronal')
plt.title('Funci贸n b')
plt.legend()

plt.tight_layout()
plt.show()
```
